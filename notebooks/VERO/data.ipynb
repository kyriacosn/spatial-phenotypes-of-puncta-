```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
x = np.linspace(-3.0, 3.0, 100)
y = np.linspace(-3.0, 3.0, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Normalize Z to have values between 0 and 1
Z_normalized = (Z - Z.min()) / (Z.max() - Z.min())

# Specifying multiple contour levels
contour_levels = np.linspace(0, 1, 11)  # 11 levels from 0 to 1

# Create a contour plot
fig, ax = plt.subplots()
CS = ax.contour(X, Y, Z_normalized, levels=contour_levels, colors='black')


plt.imshow(Z_normalized,extent = (-3,3,-3,3))
# Adding labels to each contour line
plt.clabel(CS, inline=True, fontsize=8, fmt='%1.1f')

plt.title('Contour lines with labels')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.colorbar(CS, label='Normalized Field Value')  # Add a colorbar to show the scale

# Show the plot
plt.show()
```
