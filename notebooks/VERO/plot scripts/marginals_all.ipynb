```python
import matplotlib.pyplot as plt
import json
import numpy as np
```

```python

with open("../cell1/marginals.json", 'r') as file:
    marginals1 = json.load(file)

with open("../cell2/marginals.json", 'r') as file:
    marginals2 = json.load(file)

with open("../cell3/marginals.json", 'r') as file:
    marginals3 = json.load(file)

with open("../cell4/marginals.json", 'r') as file:
    marginals4 = json.load(file)

with open("../cell5/marginals.json", 'r') as file:
    marginals5 = json.load(file)

with open("../cell6/marginals.json", 'r') as file:
    marginals6 = json.load(file)
    
marginals = [marginals1,marginals2,marginals3,marginals4,marginals5,marginals6]


with open("../cell6/marginals_lipid.json", 'r') as file:
    marginals_lip6 = json.load(file)

with open("../cell5/marginals_lipid.json", 'r') as file:
    marginals_lip5 = json.load(file)

with open("../cell4/marginals_lipid.json", 'r') as file:
    marginals_lip4 = json.load(file)

with open("../cell3/marginals_lipid.json", 'r') as file:
    marginals_lip3 = json.load(file)

with open("../cell2/marginals_lipid.json", 'r') as file:
    marginals_lip2 = json.load(file)

with open("../cell1/marginals_lipid.json", 'r') as file:
    marginals_lip1 = json.load(file)

marginals_lip = [marginals_lip1,marginals_lip2,marginals_lip3,marginals_lip4,marginals_lip5,marginals_lip6]

```

```python
fig = plt.figure(figsize=(9,4))

plot1 = plt.subplot2grid((2, 7), (0, 0), colspan=1,rowspan =1) 
plot2 = plt.subplot2grid((2, 7), (0, 1), rowspan=1, colspan=1)
plot3 = plt.subplot2grid((2, 7), (0, 2), rowspan=1, colspan=1)
plot4 = plt.subplot2grid((2, 7), (0, 3), rowspan=1, colspan=1)
plot5 = plt.subplot2grid((2, 7), (0, 4), rowspan=1, colspan=1)
plot6 = plt.subplot2grid((2, 7), (0, 5), rowspan=1, colspan=1)
plot7 = plt.subplot2grid((2, 7), (0, 6), rowspan=1, colspan=1)

first_line_plots = [plot1,plot2,plot3,plot4,plot5,plot6,plot7]

plot12 = plt.subplot2grid((2, 7), (1, 0), colspan=1,rowspan =1) 
plot22 = plt.subplot2grid((2, 7), (1, 1), rowspan=1, colspan=1)
plot32 = plt.subplot2grid((2, 7), (1, 2), rowspan=1, colspan=1)
plot42 = plt.subplot2grid((2, 7), (1, 3), rowspan=1, colspan=1)
plot52 = plt.subplot2grid((2, 7), (1, 4), rowspan=1, colspan=1)
plot62 = plt.subplot2grid((2, 7), (1, 5), rowspan=1, colspan=1)
plot72 = plt.subplot2grid((2, 7), (1, 6), rowspan=1, colspan=1)

second_line_plots = [plot12,plot22,plot32,plot42,plot52,plot62,plot72]


predictors = ["b0","BM","DM","DN","DN2","rho","sigma"]

for predictor,plot in zip(predictors,first_line_plots):
    H = 1.1*np.max([np.max(np.array(m[predictor])[:,1]) for m in marginals])

    for i,marginal in enumerate(marginals):
        plot.fill_between(np.array(marginal[predictor])[:,0], np.min(np.array(marginal[predictor])[:,1]) + i*H,  i*H + np.array(marginal[predictor])[:,1], color='blue', alpha=0.5)

    plot.plot([0,0],[-0.1*H,6*H],color = "gray",linewidth = 0.4)

    plot.set_ylim([-0.1*H,6*H])


H1 = 1.1*np.max([np.max(np.array(m["b0"])[:,1]) for m in marginals])
plot1.set_yticks(ticks = [H1/2,H1+H1/2,2*H1+H1/2,3*H1+H1/2,4*H1+H1/2,5*H1+H1/2] ,labels = ["cell1","cell2","cell3","cell4","cell5","cell6"])


for plot in first_line_plots[1:]:
    plot.set_yticks([])



for predictor,plot in zip(predictors,second_line_plots):
    H = 1.1*np.max([np.max(np.array(m[predictor])[:,1]) for m in marginals_lip])

    for i,marginal in enumerate(marginals_lip):
        plot.fill_between(np.array(marginal[predictor])[:,0], np.min(np.array(marginal[predictor])[:,1]) + i*H,  i*H + np.array(marginal[predictor])[:,1], color='blue', alpha=0.5)

    plot.plot([0,0],[-0.1*H,6*H],color = "gray",linewidth = 0.4)

    plot.set_ylim([-0.1*H,6*H])



H1 = 1.1*np.max([np.max(np.array(m["b0"])[:,1]) for m in marginals_lip])

plot12.set_yticks(ticks = [H1/2,H1+H1/2,2*H1+H1/2,3*H1+H1/2,4*H1+H1/2,5*H1+H1/2] ,labels = ["cell1","cell2","cell3","cell4","cell5","cell6"])

for plot in second_line_plots[1:]:
    plot.set_yticks([])



plot12.set_xlabel(r"$\mathrm{b_0}$")
plot22.set_xlabel(r"$\mathrm{b_{BM}}$")
plot32.set_xlabel(r"$\mathrm{b_{DM}[\mu m^{-1}]}$")
plot42.set_xlabel(r"$\mathrm{b_{DN}[\mu m^{-1}]}$")
plot52.set_xlabel(r"$\mathrm{b_{DN2}[\mu m^{-2}]}$")
plot62.set_xlabel(r"$\mathrm{\rho[\mu m]}$")
plot72.set_xlabel(r"$\mathrm{\sigma}$")

#plot1.text(-0.24,1.02,"a)",transform = plot1.transAxes, fontweight='bold')
#plot12.text(-0.24,1.02,"b) ",transform = plot12.transAxes, fontweight='bold')

plot1.text(-0.6,0.5,"peroxisomes",transform = plot1.transAxes, fontweight='bold',rotation = 90,ha = "center",va = "center")
plot12.text(-0.6,0.5,"lipid droplets",transform = plot12.transAxes, fontweight='bold',rotation = 90,ha = "center",va = "center")

plot1.set_xlim([-5,0.4])
plot12.set_xlim([-5,0.4])

plot2.set_xlim([-2.5,1.2])
plot22.set_xlim([-2.4,1.2])

plot3.set_xlim([-0.6,0.3])
plot32.set_xlim([-0.6,0.3])

plot4.set_xlim([-0.8,0.8])
plot42.set_xlim([-0.8,0.8])

plot5.set_xlim([-0.075,0.01])
plot52.set_xlim([-0.075,0.01])

plot7.set_xlim([0,2.1])
plot72.set_xlim([0,2.1])


plot6.set_xlim([0,45])
plot62.set_xlim([0,45])

plot1.set_xticks([])
plot2.set_xticks([])
plot3.set_xticks([])
plot4.set_xticks([])
plot5.set_xticks([])
plot6.set_xticks([])
plot7.set_xticks([])


plot12.set_xticks([-4,-2,0])
plot22.set_xticks([-2,-1,0])
plot32.set_xticks([-0.4,0])
plot42.set_xticks([-0.5,0,0.5],labels =[r"$-0.5$",r"$0$",r"0.5"])
plot52.set_xticks(ticks = [-0.05,0],labels =[r"$-0.05$",r"$0$"])
plot62.set_xticks([0,15,30])
plot72.set_xticks([0,1,2])


plt.subplots_adjust(wspace=0.05,hspace = 0.05)  # Adjust wspace as needed, smaller values reduce the gap
plt.savefig("../plots/posteriors.pdf")
```
